```python
%matplotlib Qt
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal
```

```python
fvz=20
dolzina=3
t = np.linspace(0, dolzina, dolzina*fvz)
s = np.where(np.sin(t*2*np.pi*2)>=0, 1, -1)

s = (t*2)%1
plt.step(t, s, 'o-')
```

```python
S = np.fft.fft(s, s.shape[0])
f_1 = np.fft.fftfreq(S.shape[0], 1/fvz) # frekvence od -fvz/2 do fvz/2
f_2 = np.linspace(0, fvz, S.shape[0], endpoint=False) # frekvence od 0 do fvz

S_rep = np.concatenate([S,]*10)
f_rep = np.linspace(0, fvz, S.shape[0], endpoint=False)
f_rep = np.concatenate([f_rep+fvz*i for i in range(-5, 5)])

cmap = plt.colormaps['Pastel1']
fig, ax = plt.subplots(2, 1, sharex=True, sharey=True)
ax[0].stem(f_rep, np.abs(S_rep), basefmt='C0', label='$[-\infty, \infty]$')
ax[0].stem(f_1, np.abs(S), markerfmt='C1', linefmt='C1', basefmt='C1', label=f'$[- {fvz//2}, {fvz//2})$')

ax[1].stem(f_rep, np.abs(S_rep), basefmt='C0')
ax[1].stem(f_2, np.abs(S), markerfmt='C2', linefmt='C2', basefmt='C2', label=f'$[0, {fvz})$')

ax[0].set_ylabel('amplituda')
ax[1].set_ylabel('amplituda')
ax[1].set_xlabel('frekvenca (Hz)')
ax[1].set_xticks(np.arange(-fvz*5, fvz*4, fvz*0.5))
ax[1].set_xlim([-fvz*1.5, fvz*1.5])

fig.legend()
fig
```
